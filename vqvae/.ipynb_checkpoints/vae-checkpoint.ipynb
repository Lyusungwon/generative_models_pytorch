```python
#model

import torch
from torch import nn
import torch.nn.functional as F

hidden_num = 400
latent_num = 10

class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.fc1 = nn.Linear(784, hidden_num)
        self.fc2 = nn.Linear(hidden_num, latent_num * 2)

    def forward(self, x):
        x = x.view(-1, 784)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = x.view(-1, 2, latent_num)
        return x

class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()
        self.fc1 = nn.Linear(latent_num, hidden_num)
        self.fc2 = nn.Linear(hidden_num, 784)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.sigmoid(self.fc2(x))
        return x
    
```

```python
#train
import torch
import dataloader
import model
import torch.utils.data
import torch.optim as optim
from torch.autograd import Variable
import torch.nn.functional as F
import torch.distributions
import sys
import numpy as np

is_cuda = torch.cuda.is_available()
# is_cuda = False

train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('/home/sungwonlyu/data/mnist', train=True, download=True,
                   transform=transforms.ToTensor()),
    batch_size=args.batch_size, shuffle=True, **kwargs)
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('/home/sungwonlyu/data/mnist', train=False, transform=transforms.ToTensor()),
    batch_size=args.batch_size, shuffle=True, **kwargs)

# dataset_test = torch.utils.data.TensorDataset(torch.FloatTensor(x_test), torch.LongTensor(y_test))#, torch.LongTensor(y_test)
# mnist_test = torch.utils.data.DataLoader(dataset_test, batch_size=10000, shuffle = True)

image_normalize = 255

prior_mean = 0
prior_std = 1

L = 10
lr = 1e-3

encoder = Encoder().cuda() if is_cuda else Encoder()
decoder = Decoder().cuda() if is_cuda else Decoder()

optimizer = optim.Adam(list(encoder.parameters())+list(decoder.parameters()), lr = lr)
epoch = 200

for i in range(epoch):
    for batch_idx, (input_data, target_data) in enumerate(train_loader):
        optimizer.zero_grad()
        batch_num = input_data.size()[0]
        input_data = input_data / image_normalize
        input_data = Variable(input_data.view(-1, 784))
        if is_cuda:
            input_data = input_data.cuda()
        params = encoder(input_data)
        z_mu = params[:, 0]
        z_sigma = torch.abs(params[:, 1])
        
        epsilon_mu = torch.zeros(batch_num, latent_num)
        epsilon_sigma = torch.ones(batch_num, latent_num)
        
        reconstruction_loss = 0
        for j in range(L):
            epsilons = torch.distributions.Normal(epsilon_mu, epsilon_sigma).sample()
            epsilons = Variable(epsilons)
            if is_cuda:
                epsilons = epsilons.cuda()
            z = z_mu + epsilons * z_sigma
            output_data = decoder(z)
            reconstruction_loss += F.binary_cross_entropy(output_data, input_data, size_average=False)
#             reconstruction_loss += torch.norm(input_data - output_data, 2)
#         kl_divergence = -0.5 + torch.log(prior_sigma) - torch.log(z_sigma) + (z_sigma^2 + (prior_means - z_mu)^2)/(2*prior_sigma^2)  
        kl_divergence = torch.sum( - torch.log(z_sigma) + (torch.pow(z_sigma, 2) + torch.pow(z_mu, 2)) / 2 )
        reconstruction_loss /= L
        loss = (reconstruction_loss + kl_divergence)/batch_num
        loss.backward()
        optimizer.step()
        print('{} epoch, Loss: {}'.format(i, loss.data[0]))
        
```
